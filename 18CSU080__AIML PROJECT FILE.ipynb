# DECISION TREE CLASSIFIER

Through the help of gdp and literacy colomn's we would be predicting wether the country belongs to latin american or sun saharan region

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()              # standard scaling the X colomn
X = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split          # train test split the data
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy')    #Fitting the data to the classifier
classifier.fit(X_train,Y_train)
```

```python
Y_pred=classifier.predict(X_test) #result prediction
```

# Performance measures 

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_pred) 
cm                     #confusion matrix
```

```python
def accuracy(cm):
    return (cm[0][0]+cm[1][1])/(cm[0][0]+cm[0][1]+cm[1][0]+cm[1][1])
accuracy(cm) #accuracy =correctly classified records(tp+tn)/wrongly classified(fp+fn)
```

```python
print("CLASSIFICATION REPORT")
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))     #classification report
```

It shows that accuracy of prediction of the latin america region is 0.8 while of sub saharan is 0.7 

```python
from sklearn import model_selection


seed = 7

kfold = model_selection.KFold(n_splits=10, random_state=seed)

model = DecisionTreeClassifier()

scoring = 'roc_auc'
# finds the area under the curve 
#more it is closer to 1 better it is
results = model_selection.cross_val_score(model, X, Y, cv=kfold, scoring=scoring)


print(results)

print(results.mean())

print(results.std())
```

As the roc_auc value is 0.8 it is an excellent fit at the dataset and the classifier does it's job
